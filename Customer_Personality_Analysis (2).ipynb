```python
# List of product categories
product_categories = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

# Create a figure and axis for each plot
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))
axes = axes.flatten() # Flatten the axes array for easy iteration

# Loop through each product category and plot
for i, category in enumerate(product_categories):
    # Group by 'Total_AcceptedCmp' and calculate mean for the current category
    data = df.groupby('Total_AcceptedCmp')[category].mean()
    data.plot(kind='bar', ax=axes[i], title=category,rot=1)
    axes[i].set_ylabel('Average Spending')
    axes[i].set_xlabel('Total Accepted Campaigns')

# Adjust layout to prevent overlap
plt.tight_layout()
```

# Creating New Columns

```python
df['Total_Under18_Home']=df['Kidhome']+df['Teenhome']
```

```python
df['Total_AcceptedCmp']=df['AcceptedCmp1']+df['AcceptedCmp2']+df['AcceptedCmp3']+df['AcceptedCmp4']+df['AcceptedCmp5']
```

# Data Cleaning

## Initial Data Clianing

```python
df['Marital_Status'].replace({'Alone':'Single','YOLO':'Single','Absurd':'Single'},inplace=True)
```

```python
df['Education'].replace({'Graduation':'Bachelor Degree','Master':'Master Degree','2n Cycle':"Master's Degree Student"},inplace=True)
```

```python
df=df[df['Income']<200000]
```

```python
df.drop(labels=['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5'], axis=1, inplace=True)
```

```python
df=df[df['Year_Birth']>=1930]
```

## Final Data Cleaning

```python
df['Total_AcceptedCmp'].replace({3:'3+',4:'3+'},inplace=True)
```

# Exporting the data

```python
df.to_csv('cleaned_data.csv')
```
